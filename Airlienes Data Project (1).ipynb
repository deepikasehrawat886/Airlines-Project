```python
## Drop unnecessary columns.
df1.drop(['Gate Location','Baggage Handling','Leg Room Service',
'Food and Drink','On-board Service','In-flight Entertainment',
'Arrival Delay','Departure Delay',
'Departure and Arrival Time Convenience','In-flight Wifi Service'],axis=1,inplace=True)
df1.head()
```

# EDA

## Age Catagory

```python
df1["Age"].unique()
```

```python
df1["Age"].nunique()
```

```python
## # making a new column for age category
Age_catagory=[]
for i in df1["Age"]:
    if i>=0 and i <20:
        Age_catagory.append("0-20")
    elif i>=20 and i<40:
        Age_catagory.append("20-40")
    elif i>=40 and i<60:
        Age_catagory.append("40-60")
    elif i>=60 and i<80:
        Age_catagory.append("60-80")
    else:
        Age_catagory.append("80+")
df1["Age_catagory"]=Age_catagory  
df1.head()
```

```python
df1["Gender"].value_counts()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
a=sns.countplot(x="Gender",data=df1,palette="rainbow_r")
for i in a.containers:
    a.bar_label(i)
```

## Here we are checking gender distribution according to age catagory:-

```python
age_gender_grp=df1.groupby(["Age_catagory","Gender"],as_index=False)["ID"].count().sort_values(by="ID",ascending=False)

age_gender_grp
```

```python
sns.set(rc={'figure.figsize':(15,5)})

a=sns.barplot(data=age_gender_grp,y='ID',x='Age_catagory',hue='Gender',palette="tab20c")
for i in a.containers:
    a.bar_label(i)
```

```python

df1['age_gen']=df1['Age_catagory']+'_'+df1['Gender']
df1.head()
```

## Here we are cheacking type of customer according to age catagory:-

```python
df1["Customer Type"].unique()
```

```python
cus_type=df1.groupby(['age_gen','Customer Type'],as_index=False)['ID'].count().sort_values(by='ID',ascending=False)
cus_type
```

```python
sns.set(rc={'figure.figsize':(15,5)})
a=sns.barplot(data=cus_type,x='age_gen',y='ID',hue='Customer Type',palette="cool")

for i in a.containers:
    a.bar_label(i)
```

```python
Travel_type=df1.groupby(['age_gen','Type of Travel'],as_index=False)['ID'].count().sort_values(by='ID',ascending=False)
Travel_type
```

```python

sns.set(rc={'figure.figsize':(15,5)})
a=sns.barplot(data=Travel_type,x='age_gen',y='ID',hue='Type of Travel',palette="winter")
for i in a.containers:
    a.bar_label(i)
```

##### AS we can see that, the both  female and male of age catagory 20 to 60 are more travelled for Business purposes.
